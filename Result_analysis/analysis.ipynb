# Result Analysis

```python
import pandas as pd
import glob
import os
```

```python
folder_results = 'C:\\Users\\Emanuele\\Desktop\\Risultati Deep Learning Strokes EXP3 Lucky split'

architecture = 'VanillaRNN' # LSTM, GRU, RNN, VanillaRNN, Attention, LNN, XLSTM, Hierarchical Attention

folder_architecture = folder_results + '\\' + architecture

# Find all CSV files containing "results" in the name
results_files = glob.glob(folder_architecture + '\\*results*.csv')

if not results_files:
    raise FileNotFoundError(f"No CSV files with 'results' in the name found in {folder_architecture}")
elif len(results_files) > 1:
    print(f"Warning: Multiple results files found: {[os.path.basename(f) for f in results_files]}")
    print(f"Using: {os.path.basename(results_files[0])}")

# Read the CSV into a dataframe
df = pd.read_csv(results_files[0])
```

```python
# group by fold and stride and get the mean of the results
df_grouped = df.groupby(['window_size', 'stride']).mean()

# sort by test accuracy
df_grouped = df_grouped.sort_values(by='test_subject_acc', ascending=False)

# drop columns that have train in the name
df_grouped = df_grouped.drop(df_grouped.columns[df_grouped.columns.str.contains('train')], axis=1)

# drop the fold column
df_grouped = df_grouped.drop('fold', axis=1)

# Save the grouped dataframe to a new CSV file in the same folder with "Test_Result" in the name
df_grouped.to_csv(folder_architecture + '\\' + "Test_Results.csv")

df_grouped
```
